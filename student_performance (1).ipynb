```python
print("\n========== NUMERIC VALIDATION ==========")

numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    print("\nColumn:", col)
    print("Min:", df[col].min())
    print("Max:", df[col].max())
    print("Mean:", df[col].mean())
    print("Zeros:", (df[col]==0).sum())
```

```python
print("\n========== CATEGORICAL VALIDATION ==========")

cat_cols = df.select_dtypes(
    include=['object','category']
).columns

print("Categorical Columns:")
print(cat_cols.tolist())


for col in cat_cols:
    print("\nColumn:", col)
    display(df[col].value_counts())
    display(df[col].value_counts(normalize=True)*100)
```

```python
print("\n========== NEGATIVE VALUES ==========")

negative_check = df.select_dtypes(include=np.number).lt(0).sum()

display(negative_check)
```

```python
print("\n========== FINAL VALIDATION REPORT ==========")

final_report = pd.DataFrame({
    "Datatype": df.dtypes,
    "Missing": df.isnull().sum(),
    "Missing %": (df.isnull().sum()/len(df))*100,
    "Unique": df.nunique()
})

display(final_report)

```

```python
import pandas as pd
import numpy as np

# Create realistic student activity data
np.random.seed(42)

data = {
    "id_student": np.arange(1001, 1021),
    "id_site": np.random.randint(1, 6, 20),
    "date": pd.date_range("2025-01-01", periods=20, freq="D"),
    "sum_click": np.random.randint(5, 100, 20)
}

df = pd.DataFrame(data)

print(df)

# Correlation
corr = df.corr(numeric_only=True)

print(corr)

# Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```
